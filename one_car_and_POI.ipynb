# 차 한대 경로와 POI 찍기
    oid만 바꿔주면서 간단하게 시각화
    시간별로 경로는 나타나지 않고 관심지점 list만 정확하게 찍혔나 확인하는 용도입니다 

```python
import pandas as pd
```

```python
df = pd.read_csv("2021_07.csv")
df
```

```python
target_id = '461000ea'
```

```python
target_id_rows = df[df['oid']=='461000ea']
target_id_rows[:100]
```

```python
# 관심지점
# location(DataFrame)
locations = pd.read_csv("K_Means_Clustering/2021_07_50_transformed.csv")

locations = locations[["lon", "lat"]]
locations.index = [f'POI{i}' for i in range(len(locations))]
```

```python
import folium

# 제주도 위도, 경도 중심 좌표
center_lat, center_lon = 33.387, 126.519

# Folium 지도 객체 생성
map = folium.Map(location=[center_lat, center_lon], zoom_start=10)


for latitude, longitude in zip(target_id_rows['latitude'], target_id_rows['longitude']):
    folium.Marker(location=[latitude, longitude]).add_to(map)

for location, row in locations.iterrows():
    latitude = row['lat']
    longitude = row['lon']
    folium.Marker(location=[latitude, longitude], icon=folium.Icon(icon='cloud',icon_color='red'),popup=location).add_to(map)


map.save("one_car_and_POI.html")
```
